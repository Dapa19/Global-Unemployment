# EDA

```python
sns.histplot(df['2022'], bins=20, kde=True)
plt.title('Unemployment Rate Distribution in 2022')
plt.xlabel('Unemployment Rate (%)')
plt.show()

sns.histplot(df['2023'], bins=20, kde=True)
plt.title('Unemployment Rate Distribution in 2023')
plt.xlabel('Unemployment Rate (%)')
plt.show()

sns.histplot(df['2024'], bins=20, kde=True)
plt.title('Unemployment Rate Distribution in 2024')
plt.xlabel('Unemployment Rate (%)')
plt.show()
```

```python
sns.pairplot(df[['2022', '2023', '2024']])
plt.show()
```

```python
# Compare unemployment rates across gender and age groups
plt.figure(figsize=(12,6))
sns.barplot(x='sex', y='2022', hue='age_group', data=df)
plt.title('Unemployment Rate by Gender and Age Group in 2022')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='sex', y='2023', hue='age_group', data=df)
plt.title('Unemployment Rate by Gender and Age Group in 2023')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='sex', y='2024', hue='age_group', data=df)
plt.title('Unemployment Rate by Gender and Age Group in 2024')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

```python
# Average unemployment rate by year
avg_unemployment_year = df[['2022', '2023', '2024']].mean()
plt.figure(figsize=(12,6))
avg_unemployment_year.plot(kind='line', marker='o')
plt.title('Average Unemployment Rate On The Last 3 Years')
plt.xlabel('Year')
plt.ylabel('Average Unemployment Rate %')
plt.show()
```
